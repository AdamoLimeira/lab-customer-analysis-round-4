```python
response_rates = df.groupby('Sales_Channel')['Response'].value_counts(normalize=True).unstack()* 100
response_rates.plot(kind='bar', stacked=True, figsize=(12, 10))
plt.show()
```

```python
response_rate = df['Response'].value_counts(normalize=True)['Yes']

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total_Claim_Amount', y=response_rate, data=df)

plt.show()
```

```python
num_no_targetvar.corr('spearman')
```

```python
multicollinearity = sns.heatmap(num_no_targetvar.corr('spearman'), annot=True)
plt.show()
```

```python
corr_matrix = num_no_targetvar.corr('spearman').abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

display(numerical_values.drop(columns = to_drop, axis=1, inplace=True))
```
